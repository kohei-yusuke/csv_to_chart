```python
import os
import csv
import numpy as np
```

```python
#@title サンプルCSVを作成するプログラム
# m行n列の行列を作成
m = 1000
n = 10

#保存場所の指定
folder_name = "csv_to_chart"
file_name = "test1.csv"
file_path = os.path.join("/content/drive/MyDrive", folder_name, file_name)

#numpy.random.normalは正規分布から値を生起する。引数は平均,分散,shapeの順
matrix = np.random.normal(0,1,(m,n))

column_names = [[str(i) for i in range(n)]]
with open(file_path, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(column_names)
    writer.writerows(matrix)

print('csvファイルが作成されました')
```
